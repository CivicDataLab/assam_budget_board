```python
import re
```

```python
grant_number = assam_exp_2018_19.groupby('GRANT NUMBER').sum()
```

```python
grant_number.reset_index(inplace = True)
```

```python
number = []
for i in grant_number.index:
    xx = grant_number['GRANT NUMBER'].iloc[i]
    r1 = re.findall(r"^\w+",xx)
    number.append(r1)
```

```python
number
```

```python
count = 0
for i in number:
    count = count + 1
```
