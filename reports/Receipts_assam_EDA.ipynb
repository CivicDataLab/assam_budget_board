## Area Wise Funds received

```python
area_receipts = pd.read_csv('/home/ubuntu/civicdatalab/Assam budgetboard/dataset/receipts data/sorted_data/sorted_Area_receipts.csv')
```

```python
fig = plt.figure(figsize = (5,5))
labels = ['General Area', 'Special Area']
sizes = area_receipts['BUDGET_2018_19']
colors = ['#ED68C1', '#F75500','#1FE04C','#19BDF7','#4A15ED','#F7F700']
explode = (0.1, 0)  # explode 1st slice
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Percentage of funds distribution between different areas(BUDGET 2018-19)')
plt.axis('equal')
plt.show

fig = plt.figure(figsize = (5,5))
labels = ['General Area', 'Special Area']
sizes = area_receipts['REVISED_2017_18']
colors = ['#ED68C1', '#F75500','#1FE04C','#19BDF7','#4A15ED','#F7F700']
explode = (0.1, 0)  # explode 1st slice
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Percentage of funds distribution between different areas(REVISED 2017-18)')
plt.axis('equal')
plt.show

fig = plt.figure(figsize = (5,5))
labels = ['General Area', 'Special Area']
sizes = area_receipts['BUDGET_2017_18']
colors = ['#ED68C1', '#F75500','#1FE04C','#19BDF7','#4A15ED','#F7F700']
explode = (0.1, 0)  # explode 1st slice
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Percentage of funds distribution between different areas(BUDGET 2017-18)')
plt.axis('equal')
plt.show


fig = plt.figure(figsize = (5,5))
labels = ['General Area', 'Special Area']
sizes = area_receipts['ACTUALS_2016_17']
colors = ['#ED68C1', '#F75500','#1FE04C','#19BDF7','#4A15ED','#F7F700']
explode = (0.1, 0)  # explode 1st slice
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Percentage of funds distribution between different areas(ACTUALS 2016-17)')
plt.axis('equal')
plt.show
```

## Funds received through different tax

```python
tax_receipts = pd.read_csv('/home/ubuntu/civicdatalab/Assam budgetboard/dataset/receipts data/sorted_data/sorted_Tax_receipts.csv')
```

```python
tax_receipts.plot(x = 'Major Head',y = ['BUDGET 2018-19','REVISED 2017-18','BUDGET 2017-18','ACTUALS 2016-17'],kind = 'bar',figsize = (50,10),color = ('#102542','#44D900','#CC1758','#F87060'))
plt.xlabel('Tax Major Heads',fontsize = 35)
plt.ylabel('Total Funds Received(Tax)',fontsize = 35)
plt.xticks(fontsize = 35,rotation = 90)
plt.yticks(fontsize = 35)
plt.title('Funds received through different taxes',fontsize = 40)
plt.legend(fontsize = 35)
#sns.pointplot(x = 'District / Block Name',y = 'Total Enrollment',data = sorted_data_Frame_2014_15)
```
